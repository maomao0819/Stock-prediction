## 2. Data Preprocessing - Preprocess the data, e.g. missing data and categorical data, etc. Explain how you deal if missing data or categorical data exist.

### missing data - Confirm no NA value.

```python
print('any NAN?', dataset.isnull().values.any())
if dataset.isnull().values.any() == True:
    dataset = dataset.dropna()
else:
    print('There is no missing data')
```

### categorical data

```python
min_0 = dataset.min() == 1
min_0_attribute = set(min_0[min_0 == True].index)
max_10 = dataset.max() == 10
max_10_attribute = set(max_10[max_10 == True].index)
continuous_variables_cols_set = set(dataset._get_numeric_data().columns)
categorical_idx = list((set(dataset.columns) - continuous_variables_cols_set) | (min_0_attribute & max_10_attribute))
print('categorical data:', categorical_idx)
```

## 3. Normalization - Would you like to consider any transformation on the data? Why or why not? Explain for each variable.

```python
std_too_big = dataset.std() > 100
std_too_big_attribute = set(std_too_big[std_too_big == True].index)
min_nonzero = dataset.min() > 0 
min_nonzero_attribute = set(min_nonzero[min_nonzero == True].index)
range_too_big = (dataset.max() - dataset.min()) > 100 
range_too_big_attribute = set(range_too_big[range_too_big == True].index)
need_log_idx = list(std_too_big_attribute & min_nonzero_attribute & range_too_big_attribute)
need_log_idx_loc = [dataset.columns.get_loc(need_log_idx[i]) for i in range(len(need_log_idx))]
need_log_idx_loc.sort()
print(need_log_idx)
```

- According to the histogram, we figure that the SalePrice, 1stFlrSF, GrLivArea in the data is skewed. 

```python
for need_log_id in need_log_idx:
    plt.title(need_log_id)
    plt.hist(dataset[need_log_id])
    plt.show()
```

### logarithm

- Take the log of the variable, and normalize it. Then, we can get the distribution of the logarithm of SalePrice, 1stFlrSF, GrLivArea.

```python
dataset_log = dataset.copy()
for need_log_id in need_log_idx:
    dataset_log[need_log_id] = np.log(dataset[need_log_id])
    dataset_log.rename(columns={need_log_id: f'log_{need_log_id}'}, inplace=True)
    plt.title(f'log_{need_log_id}')
    plt.hist(dataset_log[f'log_{need_log_id}'])
    plt.show()
dataset_log.describe()
```

### Z-score Standardize

```python
scaler = StandardScaler()
std_data = scaler.fit_transform(dataset_log)
std_dataset = pd.DataFrame(std_data, columns = dataset_log.columns)
std_dataset.describe()
```

#### (2) Stepwise regression

- Resourse: https://datascience.stackexchange.com/questions/24405/how-to-do-stepwise-regression-using-sklearn
- In order to select a better model, we do the "Forward Stepwise Regression", "Backward Stepwise Regression" and "Both Stepwise Regression".

##### A. According to the result of stepwise regression, we choose OverallQual, GrLivArea, GarageCars, TotalBsmtSF and FullBath as variables. 

```python
X = std_data.drop(std_data.columns[-1], axis = 1)
y = std_data.columns[-1]

def stepwise_selection(X, y, 
                       initial_list=[], 
                       threshold_in=0.01, 
                       threshold_out=0.05, 
                       verbose=True):
    included = list(initial_list)
    while True:
        changed=False
        # forward step
        excluded = list(set(X.columns) - set(included))
        new_pval = pd.Series(index=excluded)
        for new_column in excluded:
            model = sm.OLS(y, sm.add_constant(pd.DataFrame(X[included + [new_column]]))).fit()
            new_pval[new_column] = model.pvalues[new_column]
        best_pval = new_pval.min()
        if best_pval < threshold_in:
            best_feature = new_pval.argmin()
            included.append(best_feature)
            changed = True
            if verbose:
                print('Add  {:30} with p-value {:.6}'.format(best_feature, best_pval))

        # backward step
        model = sm.OLS(y, sm.add_constant(pd.DataFrame(X[included]))).fit()
        # use all coefs except intercept
        pvalues = model.pvalues.iloc[1:]
        worst_pval = pvalues.max() # null if pvalues is empty
        if worst_pval > threshold_out:
            changed=True
            worst_feature = pvalues.argmax()
            included.remove(worst_feature)
            if verbose:
                print('Drop {:30} with p-value {:.6}'.format(worst_feature, worst_pval))
        if not changed:
            break
    return included

result = stepwise_selection(X, y)

print('resulting features:')
print(result)
```
